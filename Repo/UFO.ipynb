```python
df_rename = df.rename(columns={'shape': 'UFO_Shape','duration (hours/min)': 'duration_Hour', 'duration (seconds)': 'duration_Seconds' })
df_rename
```

```python
df_rename['duration_Seconds'] = df_rename['duration_Seconds'].str.replace(r'\W', '', regex=True)
df_rename
```

```python
## Had to strip the ' symbol out of the duration_Seconds column ##
```

```python
df_rename['duration_Seconds'] = df_rename['duration_Seconds'].astype(float)
df_rename.info()
```

```python
## Had to convert duration seconds column to a float in order to perform a .agg and .max function int he below cells ##
```
