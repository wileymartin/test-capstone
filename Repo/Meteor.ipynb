```python
Dropped_Values = df.dropna()
```

```python
Dropped_Values['year'].astype(int)
Dropped_Values.info()
```

```python
Valid_df = Dropped_Values[Dropped_Values['nametype'] == "Valid"]
Valid_df
```

```python
### 'valid' is for most meteorites and 'relict' are for objects that were once meteorites but are now highly altered by weathering on Earth ###
```

```python
Fallen_Valid = Valid_df[Valid_df['fall'] == "Fell"]
Fallen_Valid
```

```python
## Creating a dataframe that looks at valid meteors that fell and were no found ##
```

```python
Found_Valid = Valid_df[Valid_df['fall'] == "Found"]
Found_Valid
```

```python
## Creating a dataframe that looks at valid meteors still but this time they were found ##
```

```python
## Below is a count of the type of meteors from this dataset either being valid meteors or relicts ## 
```

```python
Found_Count = Dropped_Values['nametype'].value_counts().reset_index().sort_values(by='count', ascending =False).nlargest(n=5, columns = 'count')
Found_Count.head()
```

```python
## Barplot of the counts of meteors either beign valid or relicts, as you can see there are significantly more valid counts of meteors than relicts, indicating these are more rare finds. ##
```

```python
sns.barplot(data=Found_Count, x= 'nametype', y= 'count')
```
