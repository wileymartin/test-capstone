```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
file_path = 'C:\\DA12\\Projects\\Python\\Projects\\Capstone\\test-capstone\\Data\\archive\\Meteorite_Landings.csv'

df = pd.read_csv(file_path)
```

```python
df
```

```python
df.info()
```

```python
Valid_df = df[df['nametype'] == "Valid"]
Valid_df
```

```python
### 'valid' is for most meteorites and 'relict' are for objects that were once meteorites but are now highly altered by weathering on Earth ###
```

```python
Fallen_Valid = Valid_df[Valid_df['fall'] == "Fell"]
Fallen_Valid
```

```python
Found_Valid = Valid_df[Valid_df['fall'] == "Found"]
Found_Valid
```

```python
## Below is a count of the type of meteors from this dataset either being valid meteors or relicts ## 
```

```python
Found_Count = df['nametype'].value_counts().reset_index().sort_values(by='count', ascending =False).nlargest(n=5, columns = 'count')
Found_Count.head()
```

```python
## Barplot of the counts of meteors either beign valid or relicts ##
```

```python
sns.barplot(data=Found_Count, x= 'nametype', y= 'count')
```
